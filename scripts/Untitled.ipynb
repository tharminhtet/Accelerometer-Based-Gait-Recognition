```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# loading and preprocessing data
filename = '../data/WISDM_ar_v1.1_raw.txt'
data = pd.read_csv(filename, sep=',')
data = data[data['activity'] == 'Walking']
data = data[data['user'] == 33][:100]
```

```python
# print(data[:100])
print(data.shape)
```

```python
%matplotlib inline
```

```python
plt.rcParams['figure.figsize'] = [15,10]
plt.plot(data['x-acceleration'])
plt.plot(data['y-acceleration'])
plt.plot(data['z-acceleration'])
plt.legend(['x-acceleration', 'y-acceleration', 'z-acceleration'])
```
